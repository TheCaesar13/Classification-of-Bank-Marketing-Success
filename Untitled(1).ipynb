```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as splt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import metrics 
```

```python
#i work with the reduced file

data = pd.read_csv("bank.csv", sep=";")
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
# we plot the scatter matrix plot to find relation 
pd.plotting.scatter_matrix(data)
```

```python
data.corr(method ='pearson') 
```

```python
#we have the categorical data here transform

le = preprocessing.LabelEncoder()
le.fit(data['job'])
data['job'] = (le.transform(data['job']))

le.fit(data['marital'])
data['marital'] = (le.transform(data['marital']))

le.fit(data['education'])
data['education'] = (le.transform(data['education']))

le.fit(data['default'])
data['default'] = (le.transform(data['default']))

le.fit(data['housing'])
data['housing'] = (le.transform(data['housing']))

le.fit(data['loan'])
data['loan'] = (le.transform(data['loan']))

le.fit(data['contact'])
data['contact'] = (le.transform(data['contact']))

le.fit(data['month'])
data['month'] = (le.transform(data['month']))

le.fit(data['poutcome'])
data['poutcome'] = (le.transform(data['poutcome']))

le.fit(data['y'])
data['y'] = (le.transform(data['y']))

data
```

```python
x = data.iloc[:,:-1]
x
```

```python
y = data.iloc[:,-1]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
len(X_train)
#len(X_test)
```

```python
len(X_train)
# Train
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
plot_tree(clf.fit(x,y)) 

# this clearly show that we need to focus on the feature selection.
```
