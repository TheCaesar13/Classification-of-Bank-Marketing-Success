```python
#data transformation. From categorical to numerical value
data = pd.read_csv("bank-full.csv", sep=";")

le = preprocessing.LabelEncoder()
le.fit(data['job'])
data['job'] = (le.transform(data['job']))
x = [x for x in range(len(set(data['job'])))]
print(x,le.inverse_transform(x) )

le.fit(data['marital'])
data['marital'] = (le.transform(data['marital']))

le.fit(data['education'])
data['education'] = (le.transform(data['education']))

le.fit(data['default'])
data['default'] = (le.transform(data['default']))

le.fit(data['housing'])
data['housing'] = (le.transform(data['housing']))

le.fit(data['loan'])
data['loan'] = (le.transform(data['loan']))

le.fit(data['contact'])
data['contact'] = (le.transform(data['contact']))

le.fit(data['month'])
data['month'] = (le.transform(data['month']))

le.fit(data['poutcome'])
data['poutcome'] = (le.transform(data['poutcome']))

le.fit(data['y'])
data['y'] = (le.transform(data['y']))

```

```python
###data mining and descriptive

print(data.head(),"\n",  data.describe())

### we want to know the repartition of the class
print(data.y.value_counts())

print("proportion of classes:\n", data.y.value_counts()/len(data.y))

#### here we need to plot histogram to see if there are clear boundaries to choose variable
### We need to find if there are outliers too
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
print(x.head(),"\n" ,y.head())

```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)


print("proportion of classes:\n", y_test.value_counts()/len(y_test))
###by spliting randomly we almost have the same proportion of classa 0 and 1
```

### we start apllying machine learning algorithme and we try to get the best model without overfitting

### the first algoritms are presents here and we will try to improve accuracy

```python
len(X_train)
# Train
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#plot_tree(clf.fit(x,y)) 

# this clearly show that we need to focus on the feature selection.
```

```python
confusion_ma = confusion_matrix(y_test, y_pred)

TP = confusion_ma[1, 1]
TN = confusion_ma[0, 0]
FP = confusion_ma[0, 1]
FN = confusion_ma[1, 0]

#Sensitivity
print('Sensitivity', TP / float(FN + TP))

#Specificity
print('Specificity', TN / (TN + FP))
#########################################!!! LOW SENSITIVITY!!!
```
